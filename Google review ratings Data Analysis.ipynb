```python
from sklearn.decomposition import PCA
```

```python
Performing principle component analysis to reduce the column count by the variance ratio amoung the column values
```

```python
pca = PCA()
fit_PCA = pca.fit(data_std)
fit_PCA
```

```python
pca.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt
```

```python
we need to decide how many features we’d like to keep based on the cumulative variance plot.
```

```python
plt.plot(range(len(pca.explained_variance_ratio_)), pca.explained_variance_ratio_.cumsum(), marker = 'o', linestyle = '--')
plt.figure()
```

The graph shows the amount of variance captured (on the y-axis) depending on the number of components we include (the x-axis). A rule of thumb is to preserve around 80 % of the variance. So, in this instance, we decide to keep 12 components.

```python
pca = PCA(n_components = 12)
```

```python
pca.fit(data_std)
```

calculating result components scores for the elements in our dataset

```python
scores_pca = pca.transform(data_std)
scores_pca
```

```python
from sklearn.cluster import KMeans
```

running the KMeans algorithm with a different number of clusters. Then, will determine the Within Cluster Sum of Squares or cluster convergence for each solution. Based on the values of the cluster and an approach known as the Elbow method, we make a decision about how many clusters we’d like to keep.

So testing the algorithm with 30 clusters

```python
#incorporating PCA with K-means
clusters = []
for i in range(1,31):
    kmeans_pca = KMeans(n_clusters = i, init= 'k-means++', random_state = 42)
    kmeans_pca.fit(scores_pca)
    clusters.append(kmeans_pca.inertia_)
```

The part of the graph before the elbow would be steeply declining, while the part after it – much smoother. In this instance, the kink comes at the 6 clusters mark. So, will be keeping a six-cluster solution.

```python
plt.plot(range(len(clusters)), clusters, marker = 'o', linestyle = '--')
plt.figure()
```

Here, we use the same initializer and random state as before. Subsequently, we fit the model with the principal component scores.

```python
kmeans_pca = KMeans(n_clusters = 6, init= 'k-means++', random_state = 15)
```

```python
kmeans_pca.fit(scores_pca)
kmeans_pca
```

creating a new data frame. It allows us to add in the values of the separate components to our segmentation data set. The components’ scores are stored in the ‘scores_pca’ variable. Labeling them Component 1, 2 and 3 and so on. In addition, also appending the clustering labels to the new data frame.

```python
df_pca_kmeans = pd.concat([df.reset_index(drop = True),pd.DataFrame(scores_pca)], axis = 1)
df_pca_kmeans.columns.values[-12: ] = ['Component 1','Component 2','Component 3','Component 4','Component 5','Component 6','Component 7','Component 8','Component 9','Component 10','Component 11','Component 12']
#last column for cluster designation of each row
df_pca_kmeans['Cluster_labeling'] = kmeans_pca.labels_
```

```python
df_pca_kmeans
```

creating a new column named ‘Segments’ and map the four clusters directly inside it.

```python
df_pca_kmeans['Segments'] = df_pca_kmeans['Cluster_labeling'].map({0:'first',1:'second', 2:'third', 3:'fourth', 4:'fifth',5:'sixth',6:'seventh'})
```

```python
import seaborn as sns
```

visualizing the segments with respect to the first two components as the PCA components are all independ of each other or say perpendicular and can be visualized on a 2d plane

```python
x_axis = df_pca_kmeans['Component 2']
y_axis = df_pca_kmeans['Component 1']
plt.figure(figsize = (10,8))
sns.scatterplot(x_axis,y_axis,hue = df_pca_kmeans['Segments'])
plt.title('Clusters by PCA Components')
plt.show()
```

For comparison, if we run only the k-means algorithm without the PCA step, the result would be the following:

```python
x_axis = df_pca_kmeans['Churches']
y_axis = df_pca_kmeans['resorts']
plt.figure(figsize = (10,8))
sns.scatterplot(x_axis,y_axis,hue = df_pca_kmeans['Segments'])
plt.title('Clusters by PCA Components')
plt.show()
```

In this instance, only the purple and green cluster is visually separated from the rest. The remaining three clusters are jumbled all together.

However, when we employ PCA prior to using K-means we can visually separate almost the entire data set. That was one of the biggest goals of PCA - to reduce the number of variables by combining them into bigger, more meaningful features.

```python
from sklearn.cluster import MiniBatchKMeans
from sklearn.metrics.pairwise import pairwise_distances_argmin
import numpy as np
```

Implementing the second method of clustering which is MiniBatchKmeans to visulize 

```python
mbk = MiniBatchKMeans(init ='k-means++', n_clusters = 6,
                      batch_size = 20, n_init = 10,
                      max_no_improvement = 10, verbose = 0)
```

```python
mbk.fit(scores_pca)
mbk_means_cluster_centers = np.sort(mbk.cluster_centers_, axis = 0)
mbk_means_labels = pairwise_distances_argmin(scores_pca, mbk_means_cluster_centers)
print(mbk_means_labels)
```

```python
df_pca_kmeans['MiniBatchCluster_labeling'] = mbk_means_labels
df_pca_kmeans['MiniBatchSegments'] = df_pca_kmeans['MiniBatchCluster_labeling'].map({0:'first',1:'second', 2:'third', 3:'fourth', 4:'fifth',5:'sixth',6:'seventh'})
```

```python
x_axis = df_pca_kmeans['Component 1']
y_axis = df_pca_kmeans['Component 2']
plt.figure(figsize = (10,8))
sns.scatterplot(x_axis,y_axis,hue = df_pca_kmeans['MiniBatchSegments'])
plt.title('Clusters by PCA Components')
plt.show()
```
